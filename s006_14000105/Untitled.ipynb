```python
n = 10
i = 1
while (i):
    if i<n:
#         print("{:15s}".format(i*'*'))
        print((i*'*').center(n))
        i+=2
    else:
        i-=2
        print((i*'*').center(n))
#         print(n*'*')
```

```python
for i in range(1,n+1,2):
    print((i*'*').center(n))
for i in range(n-1,0,-2):
    print((i*'*').center(n))
```
